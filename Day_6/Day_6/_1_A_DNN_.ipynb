```python
import pandas as pd
import numpy as np
import os
import nltk , re , string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from tensorflow.keras.preprocessing.text import Tokenizer
PATH='review_polarity/txt_sentoken/'
VOCAB='vocab.txt'
```

```python
# load document in memory
def load_doc(filename):
    f=open(filename)
    text=f.read()
    f.close()
    return text
```

```python
# function to clran the doucments
swords=stopwords.words('english')
def clean_doc(doc):
    tokens=word_tokenize(doc)
    tokens1=[token for token in tokens if token.isalpha]
    tokens2=[token for token in tokens if token not in swords]
    tokens3=[token for token in tokens if len(token)>1]
    return  tokens3
```

```python
clean_doc(load_doc(PATH+'neg/cv000_29416.txt'))
```

```python
# load file clean data return string
def doc_to_line(filename):
    doc=load_doc(filename)
    tokens=clean_doc(doc)
    return ' '.join(tokens)
```

```python
# load file  vocab file used in cleaning  data 
def doc_to_line_vocab(filename):
    doc=load_doc(filename)
    tokens=clean_doc(doc)
    tokens=[token for token in tokens if token in vocab_file]
    return ' '.join(tokens)
```

```python
def process_train(directory_name):
    documents=[]
    for f in os.listdir(PATH+'neg'):
        if not f.startswith('cv9'):
            docs=load_doc(directory_name+'/'+f)
            tokens=clean_doc(docs)
            documents.append(tokens)
    return documents
```

```python
tr=process_train(PATH+'neg')
len(tr)
```

```python
def process_docs(directory_name,is_train):
    documents=[]
    for f in os.listdir(directory_name):
        if is_train and  f.startswith('cv9'):
            continue
        if not is_train and not f.startswith('cv9'):
            continue
        docs=load_doc(directory_name+'/'+f)
        tokens=clean_doc(docs)
        documents.append(tokens)
    return documents
```

```python
tr=process_docs(PATH+'neg',True)
len(tr)
```

```python
# assign label to data
```

```python
def load_data(is_train):
    neg = process_docs(PATH+'neg',is_train)
    pos = process_docs(PATH+'pos',is_train)
    docs=neg+pos
    lablels=[0 for i in range(len(neg))] + [1 for i in range(len(pos))]
    return docs,lablels
```

```python
train_data,train_labels=load_data(True)
```

```python
len(train_data),len(train_labels)
```

```python
test_data,test_labels=load_data(False)
len(test_data),len(test_labels)
```

```python
# data prepration
def create_tokenizer(lines):
    tokenizer =Tokenizer()
    tokenizer.fit_on_texts(lines)
    return tokenizer
```

```python
tokenizer = create_tokenizer(train_data)
```

```python
len(tokenizer.word_index)
```

```python
x_train = tokenizer.texts_to_matrix(train_data)
x_test = tokenizer.texts_to_matrix(test_data)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from keras.models import Sequential
from keras.layers import Input , Dense
from keras.utils import to_categorical , plot_model
```

```python
# create the model
model = Sequential()
# input layer
model.add(Input(shape=(x_train.shape[1],)))
# hidden layer
model.add(Dense(256,activation='relu'))
# output layer
model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the model 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# train model
model.fit(x_train, np.array(train_labels), batch_size=5, epochs=4)
```

```python
def predict_sentiment(review):
    toknes=clean_doc(review)
    toknes=' '.join(toknes)
    encoded=tokenizer.texts_to_matrix([toknes])
    yhat=model.predict(encoded)
    prob=yhat[0,0]
    if prob > 0.5:
        return 'Positve',prob
    return 'Negative',1-prob
```
