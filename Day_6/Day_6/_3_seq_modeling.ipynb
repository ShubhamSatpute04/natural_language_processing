```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
len(alphabets)
```

```python
int_to_char = dict(enumerate(alphabets))
int_to_char
```

```python
char_to_int = dict((v,i) for i,v in enumerate(alphabets))
```

```python
char_to_int
```

```python
seq_length = 1
x = []
y = []

for i in range(len(alphabets)- seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    print(seq_in, '---->', seq_out)
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]] 
```

```python
x
```

```python
y
```

```python
import numpy as np
```

```python
x= np.reshape(x,(25 , 1, 1 ))
```

```python
x
```

```python
# normalize the data 
# if min no os 0 then divide by divide by 25 for alphabate 
x_scaled= x/25
```

```python
x_scaled
```

```python
# its a multiclass  classification problem we have to 
```

```python
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Input , SimpleRNN
```

```python
y_new=to_categorical(y)
```

```python
y_new.shape
# there will be 26 columns but here fertures are 25
```

```python
model = Sequential()

model.add(Input((1,1)))
model.add(Dense(32,activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26 , activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy' , optimizer='adam' , metrics=['accuracy'])
```

```python
model.fit(x_scaled,y_new,batch_size=1 , epochs=500)
```

```python
# give three char as input and get three char as output
seq_length = 3
x = []
y = []

for i in range(len(alphabets)- seq_length):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i+seq_length]
    print(seq_in, '---->', seq_out)
    x.append([char_to_int[char] for char in seq_in])
    y += [char_to_int[seq_out]] 
```

```python
# 
x_new = np.reshape(x, [23,seq_length,1])
```

```python
y_new = to_categorical(y)
```

```python
x_new.shape , y_new.shape
```

```python
x_new = x_new/25
```

```python
model = Sequential()

model.add(Input((3,1)))
model.add(Dense(32,activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26 , activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_new , y_new , batch_size=1 , epochs=1000)
```
