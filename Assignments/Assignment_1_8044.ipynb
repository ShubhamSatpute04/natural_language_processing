```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
sent='''They told that their ages are 23, 25 and 29 respectively.'''
```

```python
sent=word_tokenize(sent)
nums=[int(word) for word in sent if word.isdigit()]
avg=sum(nums)/len(nums)
avg
```

```python
# 2. Count total number of words starts with 't' / 'T
words=[word for word in sent if word.lower().startswith('t')]
print(words)
print(len(words))
```

```python
# 3. Count total punctuation symbols.
import string
punct=[word for word in sent if word in string.punctuation]
len(punct)
```

```python
# 4. Print the words in capital letter which ends with 'd
words=[word.upper() for word in sent if word.endswith('d')]
print(words)
```

```python
# 5. Merge all the words and print (no punctuations)
merged=''
for w in sent:
    if w not in string.punctuation:
        merged+=w
merged
```

```python
# Open the wikipedia page of Indian flag and perform following analysis

import urllib
import requests


url='https://en.wikipedia.org/wiki/Flag_of_India'
headers={
    'User-Agent':'Custom-Script/1.9 (tushar@gmail.com)',
    'Accept-Language':'en-US,en;1=0.5'
}
responce=requests.get(url,headers=headers)
responce
```

```python
# Count total number of Words
print(len(responce.text))
```

```python
# Count percentage of punctuation marks
sent=word_tokenize(responce.text)
punc=[word for word in sent if word in string.punctuation]
(len(punc)/len(responce.text))*100
```

```python
# Print total number of lines
no_lines=[word for word in sent if word=='.']
len(no_lines)
```

```python
# How many words are in title case?
titles=[]
for w in sent:
    if w not in string.punctuation and len(w) !=1 and ord(w[0])>=65 and ord(w[0])<=90:
        titles.append(w)
print(titles)
print(len(titles))
```

```python
# Remove the numbers and print the text
words=[word for word in sent if word not in string.punctuation and word.isalpha()]
print(words)
print(len(words))
```
