Bag of words


```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
```

```python
sent1='It is good practice for us'
sent2='It was also good to know about it'
tokens=word_tokenize(sent1.lower())+word_tokenize(sent2.lower())
tokens=set(tokens)
```

```python
df = pd.DataFrame([],index=[1,2],columns=list(tokens))
df
```

```python
counts1=[word_tokenize(sent1.lower()).count(word) for word in df.columns]
counts2=[word_tokenize(sent2.lower()).count(word) for word in df.columns]
```

```python
counts1
```

```python
counts2
```

```python
df.iloc[0,:]=counts1
df.iloc[1,:]=counts2
```

```python
df
```

#### N-grams
#### why
- to find frequency of words in two sentences
- Bigram
    - combination of two words
- n grams
    - combination of n words

```python
from sklearn.feature_extraction.text import CountVectorizer
cvt = CountVectorizer()
cvt=CountVectorizer(ngram_range=(1,2))
new = cvt.fit_transform([sent1,sent2])
```

```python
cvt.get_feature_names_out()
```

```python
df=pd.DataFrame(new.toarray(),columns=cvt.get_feature_names_out())
df
```

#### TF-IDE Vectorizer : 
- #### Term Frequency 
    - Its no of times it appears in a document 
    - TF=No of times t appears in a document / total no terms in the document
- #### IDF 
    -  Inverse Document Frequency of a word is the measure of how significent tem is in whole corpus
    - IDF(t) = log_e(Total number of docs / no of docs with term t in it)


```python
# Example
sent1='It was a beautiful rainy dat that made my day awesome'
sent2='We made it awesome by ading more flavors in it'
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
new=tfidf.fit_transform([sent1,sent2])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(new.toarray(),columns=tfidf.get_feature_names_out())
df
```
