```python
top=Tk()
top.geometry('800x500') #here is x not * multiply
top.title('My First Application')
l=Label(top,text='Enter the number')

def square():
    num = int(e.get()) ** 2
    result.config(text=f'Square is : {num}')

b=Button(top,text='Click Here', font=('Arial',20),command=square)
e=Entry(top)
result=Label(top,text=f'Square is')

l.pack()
e.pack()
b.pack()
result.pack()


top.mainloop()
```

```python
# display pixel 0,0 is top left corner
top=Tk()
top.geometry('800x500') #here is x not * multiply
top.title('My First Application')
l=Label(top,text='Enter the number')

def square():
    num = int(e.get()) ** 2
    result.config(text=f'Square is : {num}')

b=Button(top,text='Click Here', font=('Arial',20),command=square)
e=Entry(top)
result=Label(top,text=f'Square is')

l.place(x=100,y=50)
e.place(x=100,y=50)
b.place(x=100,y=150)
result.place(x=100,y=200)


top.mainloop()
```
