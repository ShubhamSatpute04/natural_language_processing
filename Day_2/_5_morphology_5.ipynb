```python
from polyglot.text import Word
```

```python
words=['cats','computing','association','programming','leadership','miscommunication','indentifiable','psychologically','engineering','tabular','realistic','colorfuleness','blackboard','sunflower','dashboard']
```

```python
for w in words:
    w= Word(w,language='en')
    print(w,'\t',w.morphemes)
```
