```python
from nltk.tokenize import word_tokenize
tokens=word_tokenize(text)
```

```python
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer
swords=stopwords.words('english')
import string
ls=LancasterStemmer()

tokens=[ls.stem(word.lower() )for word in tokens if word.isalpha() and word not in swords and word not in string.punctuation]
```

```python
from nltk import FreqDist
```

```python
freq=FreqDist(tokens)
```

```python
freq.plot(10)
```

```python
freq.most_common(20)
```

```python
swords=stopwords.words('english')+['retrived','archived','edit','b','original']
```

```python
url='https://en.wikipedia.org/wiki/Tiger'
response=requests.get(url,headers=headers)
soup=BeautifulSoup(response.text,'html.parser')
text=soup.get_text(strip=True)
tokens=word_tokenize(text)
token1=[ls.stem(token) for token in tokens if token.isalpha() and token.lower() not in swords and token not in string.punctuation]
freq=FreqDist(token1)
freq.plot(10)
```
