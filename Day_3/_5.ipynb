```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.wsd import lesk
```

```python
s1='I went to the bank to deposit money'
s2='We pitched our tent on the bank of the river'
```

```python
s1=lesk(word_tokenize(s1), 'bank','n')
s2=lesk(word_tokenize(s2), 'bank','n')
```

```python
s1.definition()
```

```python
s2.definition()
```
