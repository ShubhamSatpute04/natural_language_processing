```python
from nltk.corpus import wordnet

# Get all synsets for 'post'
synsets = wordnet.synsets('try')

# Loop through every synset found
for syn in synsets:
    print('---' * 10)
    print('Word and POS : ', syn.name())
    print('Meaning      : ', syn.definition())
    
    # Loop to get all synonyms (lemmas) instead of just indexing [1]
    synonyms = [lemma.name() for lemma in syn.lemmas()]
    print('Synonyms     : ', ", ".join(synonyms))
    
    # Check if examples exist before printing
    if syn.examples():
        print('Examples     : ', syn.examples())
    else:
        print('Examples     :  None available')
```

```python
for word in wordnet.synsets('bat'):
    print(word.name(), " : ", word.definition())
```

```python
for word in wordnet.synsets('bat',pos='n'):
    print(word.name(), " : ", word.definition())
```

```python
n=set()
v=set()
a=set()
for word in wordnet.synsets('sound',pos='n'):
    n.add(word.lemmas()[0].name())
for word in wordnet.synsets('sound',pos='v'):
    v.add(word.lemmas()[0].name())
for word in wordnet.synsets('sound',pos='a'):
    a.add(word.lemmas()[0].name())
print('Noun : ',n)
print('Verb : ',v)
print('Adjective : ',a)
```

```python
antonyms=set()
for word in wordnet.synsets('possible'):
    for lemma in word.lemmas():
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())
antonyms
```

```python
word=wordnet.synsets('dog')
for x in word:
   hypernyms = x.hypernyms()
   for h in hypernyms:
       print(h.name())
```

```python
word=wordnet.synsets('cat')
for x in word:
   hypernyms = x.hypernyms()
   for h in hypernyms:
       print(h.name())
```
