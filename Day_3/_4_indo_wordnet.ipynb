#### Explore the specific sunsets

```python
word= iwn.synsets('धरती')
```

```python
word[0].synset_id()
```

```python
word[0].pos()
```

```python
word[0].gloss()
```

```python
word[0].examples()
```

```python
word[0].lemma_names()
```

```python
# hyponyms
word=iwn.synsets('आंबा')
hyponyms=iwn.synset_relation(word[0],pyiwn.SynsetRelations.HYPONYMY)

for h in hyponyms:
    print(h.lemma_names()[0])
```
