```python
import os
```

```python
from pypdf import PdfReader
```

```python
for file in os.listdir():
    if file.endswith('.pdf'):
        reader=PdfReader(str(file))
        for page in reader.pages:    
            open("task_data.txt", "a").write(page.extract_text())
        
```
