```python
import spacy
import nltk
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
sent='''During the Rashtrakuta dynasty, the city was referred to as Punnaka and Punyapur. The copper plates of 758 and 768 CE show that the Yadava dynasty '''
```

```python
sent=nlp(sent)
```

```python
sent.ents
```

```python
for ent in sent.ents:
    print(ent.text,'  -->  ',ent.label_)
```
