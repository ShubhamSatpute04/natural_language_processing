```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
para='''A number is a mathematical object used to count, measure, and label. The most basic examples are the natural numbers: 1, 2, 3, 4, 5, and so forth.[1] Individual numbers can be represented in spoken or written language with number words, or with dedicated symbols called numerals; for example, "eleven" is a number word and "11" is the corresponding numeral. Viewing the concept of zero as a number required a fundamental shift in philosophy, identifying nothingness with a value. During the 19th century, mathematicians began to develop the various systems now called algebraic structures, which share certain properties of numbers, and may be seen as extending the concept. Some algebraic structures are explicitly referred to as numbers (such as the p-adic numbers and hypercomplex numbers) while others are not, but this is more a matter of convention than a mathematical distinction.[7]'''
```

```python
 ### 1.Write a program to calculate the sum of all numbers present in a paragraph.
words=word_tokenize(para)
nums=[int(word) for word in words if word.isdigit()]
print(nums)
percent=sum(nums)/len(nums)
percent
```

```python
# ### 2.
# Extract all years (4-digit numbers) from a text document and store them in a list.
year_para='''The history of modern computing spans several decades of innovation. In 1946, one of the first general-purpose computers, ENIAC, was introduced. The invention of the integrated circuit in 1958 revolutionized electronics. In 1969, ARPANET laid the foundation for the modern Internet. The first personal computers gained popularity during the late 1970s, especially in 1977. IBM launched its famous PC in 1981, changing the computer industry. The World Wide Web was proposed by Tim Berners-Lee in 1989 and became publicly available in 1991. Search engines started gaining traction around 1998 with the rise of Google. Social media platforms transformed communication after 2004, while smartphones became mainstream following the launch of the iPhone in 2007. Advances in artificial intelligence accelerated in 2012, 2016, 2020, and 2023, shaping the future of technology.'''
years=[int(word) for word in word_tokenize(year_para) if len(word)==4 and word.isdigit()]
years
```

```python
# ### 3 Find the ratio of numeric tokens to total tokens in a sentence.
words=word_tokenize(para)
nums=[word for word in words if word.isdigit()]
print(len(words),' : ',len(nums))
```

```python

# ### 4 Count how many sentences in a paragraph contain at least one number.
from nltk.tokenize import sent_tokenize
count=0
for sent in sent_tokenize(para):
    for word in word_tokenize(sent):
        if word.isdigit():
            count+=1
            break
count
```

```python
# ### 5.Tokenize a paragraph and display only words that start with a capital letter.
import re
count=[word for word in word_tokenize(para) if re.match(r'^[A-Z]', word)]
len(count)
```

```python

# ### 6.Find the top 5 most frequently occurring words after removing stopwords.
import string
from nltk.corpus import stopwords
count={}
sWords=stopwords.words('english')
words=[word.lower() for word in words]
for word in words:
    if word not in string.punctuation and word not in sWords:
       count[word]=count.get(word,0)+1
```
