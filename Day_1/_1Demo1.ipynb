```python
sent='They told that their ages are 24 31 and 29 respectively'
words=sent.split(' ')
avg=0
count=0
for l in words:
    if l.isdigit():
        avg += float(l)
        count+=1
avg=avg/count
print(avg)
```

# Custom Tokenizer
### Define by us Own

```python
import re
```

```python
def custom_tokenizer(text):
    return re.split(r"[.,;?!\s]+",text)
```

```python
text="This.is some text with punctuation > let's tokenize it. Is it ok?"
```

```python
custom_tokenizer(text)
```

```python
class CommaTokenizer:
    def tokenize(self,text):
        return re.split(r"[,\s]+",text)
```

```python
tk=CommaTokenizer()
fruits='apple,orange,manigo,graphes.'
tk.tokenize(fruits)
```
